```python
import json
import pandas as pd

path = "/home/ahmed/Desktop/AGH_Data_Agency_Holding_SA/Projet_LLM_Server/data_handling/recette/recettes_a_to_z.json"

with open(path, 'r', encoding='utf-8') as f:
    data = json.load(f)
```

```python
df_recettes = pd.json_normalize(
    data, 
    record_path=['meals']  # Assurez-vous que 'meals' est la clé contenant la liste d'objets
)
```

```python
colonnes_a_supprimer = [
    'id',
    'image',
    'youtube',
]
```

```python
df_recettes = df_recettes.drop(columns=colonnes_a_supprimer, axis=1)
```

```python
df_recettes.to_json(
    'recette_handling.json',
    orient='records',  # Ceci est la clé !
    indent=4,          # Pour une meilleure lisibilité (indentation)
    force_ascii=False  # Pour s'assurer que les caractères arabes sont correctement encodés
)
```
