```python
import pandas as pd
import json
import os
```

```python
dossier_json = '/home/ahmed/Desktop/AGH_Data_Agency_Holding_SA/Projet_LLM_Server/data_handling/suwar'
fichiers_json = [f for f in os.listdir(dossier_json) if f.endswith('.json')]
```

```python
liste_complete_de_versets = []

for nom_fichier in fichiers_json:
    chemin_complet = os.path.join(dossier_json, nom_fichier)

    with open(chemin_complet, 'r', encoding='utf-8') as f:
        data_sourate = json.load(f)

    # Extraction des données de la sourate (les données parentes)
    info_sourate = {k: v for k, v in data_sourate.items() if k != 'versets'}

    # Traitement de la liste de versets
    for verset in data_sourate['versets']:
        # Crée une nouvelle entrée en combinant les infos parentes et les infos du verset
        ligne = {**info_sourate, **verset}

        # 1. Extraire l'objet 'type' une seule fois
        type_data = ligne.pop('type')

        # 2. Utiliser la variable 'type_data' pour récupérer les valeurs
        ligne['type_fr'] = type_data['fr']
        ligne['type_ar'] = type_data['ar']
        ligne['type_en'] = type_data['en']

        # --- CORRECTION : AJOUTER LA LIGNE À LA LISTE ---
        liste_complete_de_versets.append(ligne)

# Conversion en DataFrame Pandas
df_final = pd.DataFrame(liste_complete_de_versets)

# print(f"DataFrame créé avec {len(df_final)} lignes (versets).")
# print(df_final.head())
```

```python
colonnes_a_supprimer = [
    'nom_ar',
    'nom_fr',
    'numero',
    'audio_lien',
    'ar',
    'fr',
    'type_fr',
    'type_ar',
]


df = df_final.drop(columns=colonnes_a_supprimer, axis=1)
```

```python
df = df.rename(columns={
    'nom_en': 'name',
    'type_en': 'type',
    'texte_anglais_complet': 'texte',
})
```
